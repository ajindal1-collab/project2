```python
import numpy as np
import pandas as pd
from pathlib import Path
import hvplot.pandas
```

```python
# Load the data
file_path2014 = Path('./Resources/2014.csv')
df_2014 = pd.read_csv(file_path2014, skiprows=1)[:-2]

file_path2015 = Path('./Resources/2015.csv')
df_2015 = pd.read_csv(file_path2015, skiprows=1)[:-2]

file_path2016Q1 = Path('./Resources/2016Q1.csv')
df_2016Q1 = pd.read_csv(file_path2016Q1, skiprows=1)[:-2]

file_path2016Q2 = Path('./Resources/2016Q2.csv')
df_2016Q2 = pd.read_csv(file_path2016Q2, skiprows=1)[:-2]

file_path2016Q3 = Path('./Resources/2016Q3.csv')
df_2016Q3 = pd.read_csv(file_path2016Q3, skiprows=1)[:-2]

file_path2016Q4 = Path('./Resources/2016Q4.csv')
df_2016Q4 = pd.read_csv(file_path2016Q4, skiprows=1)[:-2]
```

```python
#combined 2014 to 2016 data

df_all=[df_2014,df_2015,df_2016Q1,df_2016Q2,df_2016Q3,df_2016Q4]

raw_data = pd.concat(df_all)
raw_data.describe()
```
